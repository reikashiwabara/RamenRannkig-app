```python
# @title デフォルトのタイトル テキスト
#【GoogleDriveのファイル読み込み文】

# Google Driveをマウント
from google.colab import drive
drive.mount('/content/drive')

# pandasを使用してCSVファイルを読み込む
import pandas as pd
import os

# GoogleDriveのCSVファイルのパスを指定
file_path_tabelog = "/content/drive/MyDrive/project_assari_share/example_assari.csv"
file_path_weather = "/content/drive/MyDrive/project_assari_share/weather_data.csv"

# CSVファイルを読み込む
df_tabelog = pd.read_csv(file_path_tabelog)
df_weather = pd.read_csv(file_path_weather)

# 読み込んだデータを出力(確認のため)
print("-----csvファイルを表示-----")
print(df_tabelog)
print(df_weather)

# データフレームをリストに変換
data_list_tabelog = df_tabelog.values.tolist()
data_list_weather = df_weather.values.tolist()

# リストを出力(確認のため)
print("-----リストを表示-----")
print(data_list_tabelog)
print(data_list_weather)

# あっさり系、こってり系の分類リスト
ramen_categories = {
    'ramen case-k': 2, 'あぶらそば すずや': 2, 'どろそば屋 ひろし': 2, 'まほろば': 1,
    'らーめん 大': 2, 'パンケ 京成大久保店': 1, 'ラーメン二郎 京成大久保店': 2, '山勝角ふじ 京成大久保店': 2,
    '戯拉戯拉 京成大久保店': 1, '房総式ラーメン 人力船': 2, '日高屋 京成大久保店': 1,
    '桂林餃子 満足 京成大久保駅前店': 1, '桂香': 1, '横浜家系ラーメン 蓮': 2, '海神亭': 1,
    '長江飯店': 1, '限界を超えろ 京成大久保校': 2, '麺家 西陣': 1, '麺屋 時茂 京成大久保店': 2
}

# analysis_listを作成
analysis_list = []

for ramen in data_list_tabelog:
    name = ramen[0]
    category = ramen_categories.get(name, 0)  # デフォルトで0を返す（該当しない場合）
    analysis_list.append(ramen + [category])

# 結果を確認
for item in analysis_list:
    print(item)

#天気とらーめんの種類(あっさりorこってり)により評価値を変更する
def adjust_ramen_ratings(analysis_list, data_list_weather):
    # 天気の情報を取得（data_list_weatherの三つ目の要素が天気を示す）
    weather = data_list_weather[0][2]  # 晴れ、雨、雪など

    # analysis_listの評価を調整する
    for ramen in analysis_list:
        category = ramen[3]  # 4つ目の要素であっさり/こってりを判定
        rating = ramen[1]    # 2つ目の要素が評価値

        # 天気が晴れであっさり系（category == 1）なら評価値+0.5
        if weather == '晴れ' and category == 1:
            ramen[1] = round(rating + 0.5, 2)

        # 天気が雨や雪でこってり系（category == 2）なら評価値+0.5
        elif weather in ['雨', '雪'] and category == 2:
            ramen[1] = round(rating + 0.5, 2)

    return analysis_list

adjusted_list = adjust_ramen_ratings(analysis_list, data_list_weather)

# 結果表示
print("------評価値変更後リストの表示--------")
for ramen in adjusted_list:
    print(ramen)

#評価値順にソートする
adjusted_list = sorted(adjusted_list, key=lambda x: x[1], reverse=True)
print("------評価値順にソート後リスト表示-------------")
for ramen in adjusted_list:
    print(ramen)

#以下アップロード用コードA

# リストをDataFrameに変換
df = pd.DataFrame(adjusted_list[1:], columns=adjusted_list[0])

# Google Drive内の保存先フォルダを指定
folder_path = '/content/drive/My Drive/project_assari_share'

# CSVファイルを保存するパスを指定
csv_file_path = os.path.join(folder_path, 'analysis_data.csv')

# DataFrameをCSVファイルとして保存
df.to_csv(csv_file_path, index=False)

# アップロード完了のメッセージ
print(f"CSVファイルがGoogle Driveの {csv_file_path} に保存されました。")
```
